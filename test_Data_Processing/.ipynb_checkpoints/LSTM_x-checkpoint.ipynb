# senとsen_preの単語をIDに変換し、新たな列としてAll_dfに追加する

```python
sen = All_df['sen'].values
sen_pre = All_df['sen_pre'].values

# 単語辞書の作成
wd_set = Counter([x for s in (sen + sen_pre) for x in s])
wd_ary = np.array(list(wd_set.keys()))
wd_cnt = np.array(list(wd_set.values()))

# 出現頻度順にソート
wd_ary = wd_ary[np.argsort(wd_cnt)[::-1]]
wd_cnt.sort()
wd_cnt = wd_cnt[::-1]

# 単語ID辞書の作成
wd_to_id = {wd: i for i, wd in enumerate(wd_ary)}

# Top10の単語を出力
for i in range(10):
    print("単語:",
          list(wd_ary)[i], "\t出現数:",
          list(wd_cnt)[i], "\tID:", wd_to_id[list(wd_ary)[i]])

# 出現数CUT_OFF以下の単語のIDを統一
CUT_OFF = 2
print("words kinds:", len(wd_cnt), "words>=" + str(CUT_OFF) + ":",
      np.sum(wd_cnt >= CUT_OFF))
other_id = np.sum(wd_cnt >= CUT_OFF)
wd_to_id.update({wd: other_id for wd in wd_ary[wd_cnt < CUT_OFF]})
id_to_wd = {wd_to_id[wd]: wd for wd in wd_to_id.keys()}

# senとsen_preの単語をIDに変換
sen_id = []
sen_pre_id = []
for s, s_pre in zip(sen, sen_pre):
    sen_id.append([str(wd_to_id[wd]) for wd in s])
    sen_pre_id.append([str(wd_to_id[wd]) for wd in s_pre])

# 新し列としてAll_dfに追加
All_df.insert(loc=0, column='sen_id', value=sen_id)
All_df.insert(loc=1, column='sen_pre_id', value=sen_pre_id)
```

# word2vec を利用し、単語のベクトル辞書を作成する

```python
sen_id = All_df['sen_id'].values
sen_pre_id = All_df['sen_pre_id'].values
sen_all = np.hstack((sen_id, sen_pre_id))

max_sen_length = max([len(sen) for sen in sen_all])
print("max_sen_length:", max_sen_length)

word_vectors_size = 200

model = word2vec.Word2Vec(sen_all, size=word_vectors_size, min_count=1, window=5)
```

# 必要なデータを取り出す

```python
# データの整理（一致、重複）
print("全データ：", All_df.shape)
drop_All_df = All_df.drop_duplicates(subset=['body', 'name']).reset_index(
    drop=True)
print("重複投稿を排除したデータ：", drop_All_df.shape)

# 16types-------------------------------------
t16_All_df = drop_All_df
print("16types有効データ(重複削除)：", t16_All_df.shape)
valid = t16_All_df.shape[0]
t16_All_df = t16_All_df[t16_All_df['16types_a'] == t16_All_df[
    '16types_b']].reset_index(drop=True)
print("16types一致データ(重複削除)：", t16_All_df.shape)
print("16types一致率：", valid, "/", t16_All_df.shape[0], "=",
      t16_All_df.shape[0] / valid)
print(Counter(t16_All_df['16types_a']), "\n")
```

# データの準備
* sentences dataをpaddingし、word vectorsによりfeature vectorsを作る
* labels dataをone hotの型に変換する
* featureとlabelsをdata_helpersに転送する

```python
x_t = t16_All_df['sen_id'].values
x_t = np.array([np.array(x, dtype=np.int32) for x in x_t])
x_t_pre = t16_All_df['sen_pre_id'].values
x_t_pre = np.array([np.array(x, dtype=np.int32) for x in x_t_pre])
y_l = t16_All_df['16types_a'].values
y_l = np.array(y_l, dtype=np.int32)

a = dh.sen_to_fv(x_t, max_sen_length, model, False)
a_pre = dh.sen_to_fv(x_t_pre, max_sen_length, model, False)
b = dh.labels_to_one_hot(y_l, 16)

t16_data = dh.read_data_sets(a, a_pre, b)
```

# Model parameters

```python
embedding_dim = word_vectors_size
num_classes = 16
sequence_length = max_sen_length
num_steps = max_sen_length
# num_hidden = 800
num_lstm_hidden = 200



train_dropout = 0.8
test_dropout = 1.0

learning_rate = 0.001
batch_size = 64
total_batch = int(t16_data.train.num_examples / batch_size)
print("total_batch:", total_batch)
training_epochs = 30

X = tf.placeholder(dtype=tf.float32, shape=[None, sequence_length, embedding_dim])
X_pre = tf.placeholder(dtype=tf.float32, shape=[None, sequence_length, embedding_dim])
Y = tf.placeholder(dtype=tf.float32, shape=[None, num_classes])
keep_prob = tf.placeholder(dtype=tf.float32)

weights = {
    'out':tf.Variable(tf.truncated_normal(stddev=0.1, shape=[2*2*num_lstm_hidden, num_classes]))
}
biases = {
    'out': tf.Variable(tf.constant(value=0.1, shape=[num_classes]))
}
```

# Create model

```python
def Bi_RNN(x, x_pre, weights, biases, dropout):

    # First bi-lstm layer

    lstm_fw_cell_1 = tf.nn.rnn_cell.BasicLSTMCell(num_lstm_hidden)

    lstm_bw_cell_1 = tf.nn.rnn_cell.BasicLSTMCell(num_lstm_hidden)

    lstm_fw_cell_1_drop = tf.nn.rnn_cell.DropoutWrapper(
        lstm_fw_cell_1, output_keep_prob=dropout)
    lstm_bw_cell_1_drop = tf.nn.rnn_cell.DropoutWrapper(
        lstm_bw_cell_1, output_keep_prob=dropout)
    outputs_1, state_1 = tf.nn.bidirectional_dynamic_rnn(
        lstm_fw_cell_1_drop, lstm_bw_cell_1_drop, x, dtype=tf.float32)
    outputs_pre_1, state_pre_1 = tf.nn.bidirectional_dynamic_rnn(
        lstm_fw_cell_1_drop, lstm_bw_cell_1_drop, x_pre, dtype=tf.float32)

    output_1 = tf.concat(outputs_1, axis=2)
    output_pre_1 = tf.concat(outputs_pre_1, axis=2)

    # Second bi-lstm layer
#     lstm_fw_cell_2 = tf.nn.rnn_cell.BasicLSTMCell(num_lstm_hidden)
#     lstm_bw_cell_2 = tf.nn.rnn_cell.BasicLSTMCell(num_lstm_hidden)

#     lstm_fw_cell_2_drop = tf.nn.rnn_cell.DropoutWrapper(
#         lstm_fw_cell_2, output_keep_prob=dropout)
#     lstm_bw_cell_2_drop = tf.nn.rnn_cell.DropoutWrapper(
#         lstm_bw_cell_2, output_keep_prob=dropout)
#     outputs_2, state_2 = tf.nn.bidirectional_dynamic_rnn(
#         lstm_fw_cell_2_drop, lstm_bw_cell_2_drop, output_1, dtype=tf.float32)
#     outputs_pre_2, state_pre_2 = tf.nn.bidirectional_dynamic_rnn(
#         lstm_fw_cell_2_drop, lstm_bw_cell_2_drop, output_pre_1, dtype=tf.float32)

#     output_2 = tf.concat(outputs_2, axis=2)
#     output_pre_2 = tf.concat(outputs_pre_2, axis=2)

    # Full connection layer
    lstm_out = tf.reduce_mean(output_1, axis=1)
    lstm_out_pre = tf.reduce_mean(output_pre_1, axis=1)

    lstm_out_concat = tf.concat([lstm_out, lstm_out_pre], axis=1)

    out = tf.add(tf.matmul(lstm_out_concat, weights['out']), biases['out'])
    return out
```

# Construct and Evaluate model

```python
y_pred = Bi_RNN(X, X_pre, weights, biases, keep_prob)
y_softmax = tf.nn.softmax(y_pred)

# Define loss and optimizer
# type 1(old):
# loss = tf.reduce_mean(
#     -tf.reduce_sum(Y * tf.log(y_softmax), reduction_indices=[1]))
# type 2(server):
loss = tf.reduce_mean(
    tf.nn.softmax_cross_entropy_with_logits(labels=Y, logits=y_pred))
# type 3(new):
# loss = tf.reduce_mean(
#     tf.nn.softmax_cross_entropy_with_logits_v2(labels=Y, logits=y_pred))
optimizer = tf.train.AdamOptimizer(learning_rate).minimize(loss)

# Evaluate model
pred = tf.argmax(y_pred, 1)
true = tf.argmax(Y, 1)
correct_prediction = tf.equal(pred, true)
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
```

# Training and Display

```python
sess = tf.InteractiveSession()
tf.global_variables_initializer().run()

# Training cycle
all_x_test = t16_data.test.vectors
all_x_test_pre = t16_data.test.vectors_pre
all_y_test = t16_data.test.labels
start = time.time()
for epoch_i in range(training_epochs):
    ave_loss = 0
    for batch_i in range(total_batch):
        batch_x, batch_x_pre, batch_y = t16_data.train.next_batch(batch_size)
        _, l = sess.run(
            [optimizer, loss],
            feed_dict={
                X: batch_x,
                X_pre: batch_x_pre,
                Y: batch_y,
                keep_prob: train_dropout
            })
        ave_loss += l / total_batch
    # Display logs per epoch step
    if epoch_i % 1 == 0:
        train_acc = sess.run(
            accuracy,
            feed_dict={
                X: batch_x,
                X_pre: batch_x_pre,
                Y: batch_y,
                keep_prob: train_dropout
            })
        test_acc = sess.run(
            accuracy,
            feed_dict={
                X: all_x_test,
                X_pre: all_x_test_pre,
                Y: all_y_test,
                keep_prob: test_dropout
            })
        print("Epoch:%2d Batch:%4d" % (epoch_i + 1, batch_i + 1),
              "Train_acc=[%.3f]" % train_acc, "Test_acc=[%.3f]" % test_acc,
              "Train_loss(ave)=[%5.3f]" % ave_loss)
end = time.time()
print("Process Time :%.2f s" % (end - start))
sess.close()
```
